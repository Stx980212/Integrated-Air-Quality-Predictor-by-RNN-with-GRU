```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
years = [2016,2017]
data = pd.DataFrame()
```

## Weather Data

```python
weather = pd.DataFrame()
for y in years:
    temp = pd.read_csv("%i_12_weather.csv"%y)
    weather = weather.append(temp)
weather = weather.reset_index()
weather = weather.drop(columns = ["index"])
```

```python
data = weather.copy()
weather.head()
```

```python
weather.columns
```

```python
data["date"] = pd.to_datetime(data["date"])
data["Year"] = data["date"].dt.year
data["Month"] = data["date"].dt.month
data["Day"] = data["date"].dt.day
```

## Solar Data

```python
solar = pd.DataFrame()
for y in years:
    temp = pd.read_csv("83556_34.05_-118.26_%i.csv"%y)
    solar = solar.append(temp)
solar = solar.reset_index()
solar = solar.drop(columns = ["index"])
```

```python
data["GHI"] = solar["GHI"]
data["Clearsky GHI"] = solar["Clearsky GHI"]
solar.head()
```

## AQI Data

### PM2.5

```python
PM = pd.DataFrame()
for y in years:
    temp = pd.read_csv("%i_88101_4008.csv"%y)
    PM = PM.append(temp)
PM = PM.reset_index()
PM = PM.drop(columns = ["index","level_0","Unnamed: 0"])
```

```python
PM["sample_measurement"] = PM["sample_measurement"].fillna(PM["sample_measurement"].interpolate())
data["PM2.5_obj"] = PM["sample_measurement"]
PM.head()
```

### NOx

```python
NOx = pd.DataFrame()
for y in years:
    temp = pd.read_csv("%i_42602_4008.csv"%y)
    NOx = NOx.append(temp)
NOx = NOx.reset_index()
NOx = NOx.drop(columns = ["index","level_0","Unnamed: 0"])
```

### SO2

```python
SO2 = pd.DataFrame()
for y in years:
    temp = pd.read_csv("%i_42401_4006.csv"%y)
    SO2 = SO2.append(temp)
SO2 = SO2.reset_index()
SO2 = SO2.drop(columns = ["index","level_0","Unnamed: 0"])
```

```python
DateTime = NOx[["date_local","time_local"]]
SO2_alt = pd.merge(DateTime,SO2,on = ["date_local","time_local"], how = "left")
SO2_alt["sample_measurement"] = SO2_alt["sample_measurement"].fillna(SO2_alt["sample_measurement"].interpolate())
```

```python
data["SO2_obj"] = SO2_alt["sample_measurement"]
SO2_alt.head()
```

```python
data_t = pd.merge(data,consumption,on = ["Year","Month"], how = "left")
```

## Traffic

```python
traffic = pd.read_csv("Traffic_Collision_Data_from_2010_to_Present.csv")
```

```python
traffic["date"] = pd.to_datetime(traffic["Date Occurred"])
traffic["Year"] = traffic["date"].dt.year
traffic["Month"] = traffic["date"].dt.month
traffic["Day"] = traffic["date"].dt.day
```

```python
traffic_count = traffic.groupby(["Year","Month","Day"])["DR Number"].count()
```

```python
data_all = pd.merge(data_t,traffic_count,on = ["Year","Month","Day"], how = "left")
data_all = data_all.rename(columns={'DR Number':'Accident Count'})
```

```python
data_all["Hour"] = data_all["time"] / 100
data_all["Hour"] = data_all["Hour"].astype(int)
```

```python
data_all.to_csv("Data.csv")
```
